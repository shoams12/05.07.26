```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
import sympy as sp
```

```python
#Task 1
#Draw the lines for:
#x + y = 12 -> y = 12 - x
#x - y = 4 -> y = x - 4
#Add grid, legend, and axis lines

x_values = np.linspace(-15, 35, 20)
y1_values = 12 - x_values
y2_values = x_values - 4

plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # x-axis
plt.plot(x_values, y1_values, color='blue', label = 'y = 12 - x')
plt.plot(x_values, y2_values, color='orange', label = 'y = x - 4')
plt.legend()
plt.title('12 - x = x - 4')
plt.xlim(-10, 20)
plt.ylim(-20, 20)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

```python
#Task 2
#Use symbols, Eq, and solve to find the intersection point. Scatter the point and annotate it on the graph
x = sp.symbols('x')

eq1 = 12 - x     
eq2 = x - 4       

solution = sp.solve( sp.Eq(eq1, eq2), x )
x_point = float(solution[0])

y_point = 12 - x_point

print("Intersection point:", (x_point, y_point))

x_values = np.linspace(-15, 35, 20)

y1 = 12 - x_values 
y2 = x_values - 4

plt.figure(figsize=(8, 5))
plt.plot(x_values, y1, label='y = 12 - x')
plt.plot(x_values, y2, label='y = x - 4')

plt.scatter(x_point, y_point, s=100, color = 'blue', zorder=5)

plt.annotate(
    f'({x_point:.1f}, {y_point:.1f})',
    (x_point, y_point),
    xytext=(10, 10),
    textcoords='offset points'
)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Intersection Point of Two Lines')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Task 3
#The label points are Y = np.array([2, 4, 8, 20])
#In the graph, the predicted Y values were Y_pred = np.array([2, 5, 10, 25])
#Calculate the MSE

Y = np.array([2, 4, 8, 20])

Y_pred = np.array([2, 5, 10, 25])

mse = np.mean((Y - Y_pred) ** 2)

print("MSE:", mse)
```
